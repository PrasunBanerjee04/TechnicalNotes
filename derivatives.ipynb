```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import scipy.stats as ss
```

```python
# plotting forex hedge forward payoff 

def forex_forward_payoff(r_market, r_fixed, amount):
    pnl = amount * (r_market - r_fixed)
    return pnl

r_fixed = 1.2234 
amount = 1000000
x = np.linspace(0.5, 2, 100)
y = [forex_forward_payoff(x_i, r_fixed, amount) for x_i in x]
plt.plot(x,y)
plt.axhline(0.0, color='black', linestyle=':')
plt.vlines(r_fixed, ymin=min(y), ymax=0, color='r', linestyle=':')
plt.xlabel('$S_T$')
plt.ylabel('PnL')
plt.title("Payoff from Long Forward Contract")
```
