```python
# valuate a function defined as y = f(x) = sign(sin(2πx))x^2 when x ≥ 0 and y = f(x) = sign(cos(2πx))x^3 + x when x < 0

def f(x):

    def sign(num):
        if num > 0:
            return 1
        elif num < 0:
            return -1
        else:
            return 0
        
    if x < 0:
        s = sign(np.cos(2 * np.pi * x))
        return s * x**3 + x
    else:
        s = sign(np.sin(2 * np.pi * x))
        return s * x**2

f(2)
```

```python
# plotting a surface 

def f(x,y):
    return np.log(np.cos(y) / np.sin(x))

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
x = np.linspace(-np.pi / 2, np.pi / 2, 50)
y = np.linspace(-np.pi / 2, np.pi / 2, 50)
X, Y = np.meshgrid(x, y)
Z = f(X,Y)
ax.plot_surface(X, Y, Z, cmap="viridis")
plt.show()
```
